```python
mask = ['gender' , 'ever_married' , 'work_type' , 'Residence_type' , 'smoking_status']
```

```python
data_feature = df[mask]
data_feature.head()
```

```python
df = df.drop(['gender' , 'ever_married' , 'work_type' , 'Residence_type' , 'smoking_status','id','stroke'],axis=1)
df.head()
```

# Chuyển đổi dữ liệu của data_feature

```python
mapper_gender = {'Male':0 , 'Female':1 , 'Other':2}
data_feature['gender'] = data_feature['gender'].map(mapper_gender)
data_feature.head()
```

```python
mapper_married = {'Yes':1 , 'No':0}
data_feature['ever_married'] = data_feature['ever_married'].map(mapper_married)
data_feature.head()
```

```python
print(set(data_feature['work_type']))
```

```python
mapper_work = {'Private':0 , 'Govt_job':1 , 'Self-employed':2 , 'Never_worked':3 , 'children':4}
data_feature['work_type'] = data_feature['work_type'].map(mapper_work)
data_feature.head()
```

```python
mapper_residence = {'Urban':1 , 'Rural':0}
data_feature['Residence_type'] = data_feature['Residence_type'].map(mapper_residence)
data_feature.head()
```

```python
print(set(data_feature['smoking_status']))
```

```python
mapper_smoking = {'formerly smoked':0 , 'Unknown':1 , 'smokes':2 , 'never smoked':3}
data_feature['smoking_status'] = data_feature['smoking_status'].map(mapper_smoking)
data_feature.head()
```

```python
df_final = pd.concat([df,data_feature],axis=1)
df_final.head()
```

```python
df_final.info()
```

# Chia nhỏ dữ liệu để train mô hình máy học

```python
# x là dữ liệu để train mô hình
x = df_final
x.head()
```

```python
df_original = pd.read_csv(path)
# y là dữ liệu để check kết quả dự đoán
y = df_original['stroke']
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train , x_valid , y_train , y_valid = train_test_split(x,y,train_size=0.7 , test_size=0.3 , random_state=0)
```

```python
x_train.shape
```

```python
x.shape
```

# training mô hình máy học

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_train , y_train)
```

```python
y_predict = model.predict(x_valid.head())
y_predict  
```
